```python
top_10_goal_scorers = df[df['goals'] > 0].sort_values('goals', ascending=False).head(10)
print(top_10_goal_scorers[['player_name', 'goals']])
```

```python
top_10_longest_playing_players = df[df['time'] > 0].sort_values('time', ascending=False).head(10)
print(top_10_longest_playing_players[['player_name', 'time']])
```

```python
top_10_assists_leaders = df[df['assists'] > 0].sort_values('assists', ascending=False).head(10)
print(top_10_assists_leaders[['player_name', 'assists']])
```

```python
top_4_mancity_goal_scorers = df[(df['team_title'] == 'Manchester City') & (df['goals'] > 0)].sort_values('goals', ascending=False).head(4)
print(top_4_mancity_goal_scorers[['player_name', 'goals']])
```

```python
team_goals = df.groupby('team_title')['goals'].sum()
most_goal_scorers_team = team_goals.idxmax()
print(most_goal_scorers_team)
```

```python
top_assist = df[df['assists'] > 0].sort_values('assists', ascending=False).head(1)
yellow_cards = top_assist['yellow_cards'].values[0]
red_cards = top_assist['red_cards'].values[0]
print(f"Yellow cards: {yellow_cards}")
print(f"Red cards: {red_cards}")
```

```python
top_player = df[(df['assists'] > 0) & (df['goals'] > 0)].sort_values(['assists', 'goals'], ascending=False).head(1)
print(top_player[['player_name', 'team_title', 'assists', 'goals']])
```

```python
team_goals = df.groupby('team_title')['goals'].sum()
top_5_goal_scoring_teams = team_goals.nlargest(5)
print(top_5_goal_scoring_teams)
```

```python
no_cards = df[(df['yellow_cards'] == 0) & (df['red_cards'] == 0)]
no_cards_goals = no_cards.sort_values('goals', ascending=False)
print(no_cards_goals[['player_name', 'goals']])
```
